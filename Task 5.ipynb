1) Create a function getting two integer inputs from user, and print the following : a)Addition of two numbers is +value. b)Subtraction of two numbers is +value. c)Division of two numbers is +value. d)Multiplication of two numbers is +value.

```python
num1 = int(input("Enter First Number :"))
num2 = int(input("Enter Second Number :"))
print("First Number is :",num1)
print("Second Number is :",num2)
def arithmetic(num1,num2):
    print("Addition of two numbers is :",num1 + num2)
    print("Subtraction of two numbers is :",num1 -num2)
    print("Division of two numbers is :",num1 / num2)
    print("Multiplication of two numbers is :",num1 * num2)
arithmetic(num1,num2)
```
